```python
# premierèment on aura à importer les bibliothèques 

import numpy as np
import pandas as pd 
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error
```

```python
# on va charger nos données iris 
iris = datasets.load_iris()
X = iris.data # caracterisiques ou les input
y = iris.target # cible ou label
```

```python
#  preparation des données 
y = ( y== 0 ).astype(int) # conversion en binaire
```

```python
# en suite on aura à diviser nos données 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train,X_test, y_train, y_test
```

# on aura à entrainer nos modèles

```python
 # modèle ridge

ridge_model = Ridge(alpha = 1.0 ) # paramètre de regularisation
ridge_model.fit(X_train, y_train)
```

```python
# modele Lasso
lasso_model = Lasso(alpha = 0.1)
lasso_model.fit(X_train, y_train)
```

```python
# en suite on fera la prédiction et evaluer les performences de nos modèles 
ridge_predictions = ridge_model.predict(X_test)
lasso_predictions = lasso_model.predict(X_test)
```

```python
# en fin on aura à calculer l'erreur quadratique moyenne
ridge_rmse = np.sqrt(mean_squared_error(y_test,ridge_predictions.round()))
lasso_rmse = np.sqrt(mean_squared_error(y_test,lasso_predictions.round()))
```

```python
# afficher les resultats
print(f"Ridge RMSE: {ridge_rmse:2f}")
print(f"Lasso RMSE: {lasso_rmse:4f}")
```

on peut dire que :

```python
# dans nos deux methodes on canstante que le ROOT MEAN SQUARED ERROR est 0.00000 par tout 
#cela nous indiquer que le modele s'ajuste parfaitement aux données d'entrainement 
# et donc nos données sont simples
```
